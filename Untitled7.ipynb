```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

LOAD DATA

```python
employees=pd.read_csv("C:\\Users\\vky97\\OneDrive\\Desktop\\employees.csv")
departments=pd.read_csv("C:\\Users\\vky97\\OneDrive\\Desktop\\departments.csv")
jobs=pd.read_csv("C:\\Users\\vky97\\OneDrive\\Desktop\\jobs.csv")
```

Merge Data For Analysis

```python
df=employees.merge(departments,  on="DepartmentID", how='left')\
            .merge(jobs, on='JobID', how='left')
```

Sample Analysis Ideas

a. Total employees per department

```python
plt.figure(figsize=(12,6))
dept_count=df['DepartmentName'].value_counts()
plt.plot(dept_count.index,dept_count.values, color='red', marker='o')
plt.xticks(rotation=45)
plt.title("Total Employee Per department")
plt.ylabel("Total Employee")
plt.show()
```

b. Male vs Female distribution

```python
gend_count=df['Gender'].value_counts()
plt.bar(gend_count.index,gend_count.values, color=['red','blue'], label="jdkfs")
plt.title("Gender Distribution")
plt.show()
```


C.   Employees per department

```python
plt.figure(figsize=(10,6))
sns.countplot(x="DepartmentName", data=df, color='g', order=df['DepartmentName'].value_counts().index)
plt.xticks(rotation=45)
plt.ylim(1,80)
plt.title('Employees per Department')
# plt.tight_layout()
plt.show()
```

D. d. Salary range by job title (boxplot)


```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['JobTitle'], y=(df['MinSalary']+df['MaxSalary'])/2)
plt.xticks(rotation=45)
plt.title("Salary Distribution By Job Title")
plt.ylabel("Salary")
plt.show()
```

e. Gender percentage per department

```python
gender_dept = df.groupby(['DepartmentName','Gender']).size().unstack()
gender_dept['Female_Percentage'] = gender_dept['Female'] / gender_dept.sum(axis=1) * 100
gender_dept[['Female_Percentage']].plot(kind='bar', figsize=(10,5))
plt.title('Female Percentage per Department')
plt.xticks(rotation=45)
plt.show()
```

f. Hires over time

```python
import matplotlib.pyplot as plt

# Hire date ko proper datetime me convert karna
df['HireDate'] = pd.to_datetime(df['HireDate'], dayfirst=True, errors='coerce')

# Group by year
hires_per_year = df.groupby(df['HireDate'].dt.year).size()

# Plot
plt.figure(figsize=(10,6))
plt.plot(hires_per_year.index, hires_per_year.values, 
         marker='o', markersize=10, linewidth=2, color='royalblue')

# Title & Labels
plt.title('Employees Hired per Year', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Hires', fontsize=12)

# Grid add karna
plt.grid(True, linestyle='--', alpha=0.6)

# Bar ke upar values show karna
for x, y in zip(hires_per_year.index, hires_per_year.values):
    plt.text(x, y+1, str(y), ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
